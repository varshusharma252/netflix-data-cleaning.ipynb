```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
duplicates = df.duplicated().sum()
print(duplicates)
```

```python
if df.duplicated().any():
    df.drop_duplicates(inplace=True)
    print("There are duplicates in the dataset.")
else:
    print("There are no duplicates in the dataset.")
```

```python
nullcount = df.isnull().sum()
print(nullcount)
```

```python
df['director'].fillna("Unknown", inplace=True)
df['cast'].fillna("Unknown", inplace=True)
df['country'].fillna("Unknown", inplace=True)
df['rating'].fillna("Unknown", inplace=True)
df['date_added'].fillna("Unknown", inplace=True)
```

```python
nullcountafterfix = df.isnull().sum()
print(nullcountafterfix)
```

```python
df['type'] = df['type'].str.strip().str.title()
df['country'] = df['country'].str.strip().str.title()
df['rating'] = df['rating'].str.strip()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.to_csv("netflix_cleaned.csv", index=False)
```

Summary of Data Cleaning (Netflix Titles Dataset)

In this project, the Netflix dataset was cleaned and prepared using Python (Pandas).
The following data cleaning steps were performed:

Removed duplicate rows using df.drop_duplicates().

Handled missing values:

Filled missing values in director, cast, country, rating, and date_added with "Unknown".

Cleaned text formats:

Standardized type and country using .str.strip().str.title().

Cleaned rating values using .str.strip().

Converted the date_added column to datetime format using pd.to_datetime().

Renamed column headers to lowercase and replaced spaces with underscores.

Checked missing values and dataset info using df.info() and df.isnull().sum().

Exported the cleaned dataset as netflix_cleaned.csv
